Q4-a

```python
from PIL import Image
import numpy as np

def img_to_array(path):
    # Load the image
    img = Image.open(path)

    # Check if the image is in grayscale or RGB
    if img.mode == 'L':  # Grayscale
        img_array = np.array(img)
        mode = 'L'
    else:  # RGB
        img_array = np.array(img)
        mode = 'RGB'

    # Save the array to a text file with dimensions
    file_name = path.split('/')[-1].split('.')[0] + '_array.txt'
    with open(file_name, 'w') as f:
        f.write(f'{mode}\n')
        f.write(f'{img_array.shape}\n')
        np.savetxt(f, img_array.flatten(), fmt='%d')

    print(f'Image array saved to {file_name}')
    return file_name
```

Q4-b

```python
def load_image_array(file_name):
    with open(file_name, 'r') as f:
        mode = f.readline().strip()
        shape = tuple(map(int, f.readline().strip()[1:-1].split(',')))
        img_array = np.loadtxt(f, dtype=int)

    # Reshape the array based on the stored dimensions
    img_array = img_array.reshape(shape)

    return img_array
```

```python
# Save the image array to a text file
file_name = img_to_array('ss1.jpg')

# Load the image array from the text file
img_array = load_image_array(file_name)

# Display the image (for verification)
import matplotlib.pyplot as plt

if img_array.ndim == 2:  # Grayscale
    plt.imshow(img_array, cmap='gray')
else:  # RGB
    plt.imshow(img_array)
plt.show()
```
