# Q1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
```

```python
url = 'https://timesofindia.indiatimes.com/home/headlines'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

for i, headline in enumerate(soup.find_all('a', attrs={'class': 'w_img'})[:10], 1):
    title_tag = soup.find('title')
    title = title_tag.get_text(strip=True)
    link = headline['href']
    print(f"Title {i}: {title}")
    print(f"Link {i}: https://timesofindia.indiatimes.com{link}")
```

# Q2

```python
search_query = "software engineer"
location = "Delhi"
url = f"https://www.linkedin.com/jobs/search/?keywords={search_query}&location={location}"

headers = {
    'User-Agent': 'Mozilla/5.0'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
jobs = soup.find_all('div', class_='base-card')

for job in jobs:
    title = job.find('h3', class_='base-search-card__title').get_text(strip=True)
    company = job.find('h4', class_='base-search-card__subtitle').get_text(strip=True)
    link = job.find('a', class_='base-card__full-link')['href']

    print(f"Job Title: {title}")
    print(f"Company: {company}")
    print(f"Link: {link}\n")
```
