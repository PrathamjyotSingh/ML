```python

def search_symbol(company_name, api_key):
    url = f"https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={company_name}&apikey={api_key}"
    response = requests.get(url)
    data = response.json()
    return data['bestMatches']

api_key = '7N40Z62VR98CFE51'

# Search for Apple
apple = search_symbol("Apple", api_key)
print("Apple:", apple)

videocon = search_symbol("Videocon", api_key)
print("Videocon:", videocon)

intel = search_symbol("Intel Corporation", api_key)
print("Intel Corporation:", intel)
```

```python

def get_intraday_data(symbol, interval, api_key):
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval={interval}&outputsize=full&apikey={api_key}"
    response = requests.get(url)
    data = response.json()

    time_series_key = f"Time Series ({interval})"
    time_series_data = data[time_series_key]

    df = pd.DataFrame.from_dict(time_series_data, orient="index")
    df = df.rename(columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",
        "4. close": "Close",
        "5. volume": "Volume"
    })

    df.index = pd.to_datetime(df.index)
    return df


api_key = '7N40Z62VR98CFE51'

apple_intraday_data = get_intraday_data("AAPL", "15min", api_key)

print(apple_intraday_data)
```

```python
import requests
import json


api_key = 'VK0HGERQKXSY33QV'
videocon_symbol = '511389.BSE'

url = 'https://www.alphavantage.co/query'
params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': videocon_symbol,
    'outputsize': 'compact',
    'apikey': api_key
}

response = requests.get(url, params=params)

print("Raw Response:")
print(response.text)

if response.status_code == 200:
    data = response.json()

    time_series = data.get("Time Series (Daily)", {})

    if not time_series:
        print("No daily data available or symbol might be incorrect.")
    else:
        print("\nDaily Data:")
        for date, values in time_series.items():
            print(f"Date: {date}")
            print(f"  Open: {values['1. open']}")
            print(f"  High: {values['2. high']}")
            print(f"  Low: {values['3. low']}")
            print(f"  Close: {values['4. close']}")
            print(f"  Volume: {values['5. volume']}\n")
```

```python
import requests

api_key = '7N40Z62VR98CFE51'
ticker = 'IBM'
topic = 'technology'

url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&topics={topic}&apikey={api_key}'
response = requests.get(url)
data = response.json()

if 'feed' in data:
    articles = data['feed']
    for article in articles:
        print(f"Title: {article['title']}")
        print(f"Summary: {article['summary']}")
        print(f"Authors: {', '.join(article.get('authors', []))}")
        print(f"Overall Sentiment Score: {article['overall_sentiment_score']}")
        print(f"Overall Sentiment Label: {article['overall_sentiment_label']}")
        print("-" * 40)
else:
    print("Error retrieving data:", data.get("Error Message", "Unknown error"))
```

```python
import requests
import yfinance as yf

api_key = '7N40Z62VR98CFE51'
ticker = 'IBM'

def fetch_annual_income_statement(api_key, ticker):
    annual_url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'

    try:
        response = requests.get(annual_url)
        response.raise_for_status()
        data = response.json()

        if 'annualReports' in data:
            print(f"\nAnnual Income Statement for {ticker}:")
            for report in data['annualReports']:
                print(f"Total Revenue: {report.get('totalRevenue', 'N/A')}")
                print(f"Cost of Revenue: {report.get('costOfRevenue', 'N/A')}")
                print(f"Gross Profit: {report.get('grossProfit', 'N/A')}")
                print(f"Operating Income: {report.get('operatingIncome', 'N/A')}")
                print(f"Net Income: {report.get('netIncome', 'N/A')}")
                print("-" * 40)
        else:
            print("Error retrieving annual income statement data:", data.get("Error Message", "Unknown error"))

    except requests.RequestException as e:
        print(f"An error occurred: {e}")

def fetch_quarterly_income_statement(ticker):
    try:
        stock = yf.Ticker(ticker)
        quarterly_financials = stock.quarterly_financials

        print(f"\nQuarterly Income Statement for {ticker}:")
        print(quarterly_financials)

    except Exception as e:
        print(f"An error occurred: {e}")

fetch_annual_income_statement(api_key, ticker)
fetch_quarterly_income_statement(ticker)
```

```python
import requests

# API key and URL
api_key = '7N40Z62VR98CFE51'
url = f'https://www.alphavantage.co/query?function=CURRENCY_EXCHANGE_RATE&from_currency=USD&to_currency=INR&apikey={api_key}'

try:
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    if 'Realtime Currency Exchange Rate' in data:
        exchange_rate = data['Realtime Currency Exchange Rate']
        print(f"Real-time exchange rate for USD to INR:")
        print(f"From Currency Code: {exchange_rate.get('1. From_Currency Code', 'N/A')}")
        print(f"From Currency Name: {exchange_rate.get('2. From_Currency Name', 'N/A')}")
        print(f"To Currency Code: {exchange_rate.get('3. To_Currency Code', 'N/A')}")
        print(f"To Currency Name: {exchange_rate.get('4. To_Currency Name', 'N/A')}")
        print(f"Exchange Rate: {exchange_rate.get('5. Exchange Rate', 'N/A')}")
        print(f"Last Refreshed: {exchange_rate.get('6. Last Refreshed', 'N/A')}")
        print(f"Time Zone: {exchange_rate.get('7. Time Zone', 'N/A')}")
    else:
        print("Error retrieving exchange rate data:", data.get("Error Message", "Unknown error"))

except requests.RequestException as e:
    print(f"An error occurred: {e}")
```
